```python
import os, sys
sys.path.append("/home/centos/CGMD/")
from pyMD.file_parser import LmpParser
from pyMD import collective_structure_class as csc
import os
```

```python
def remove_period(atom3d:csc.Atom3D, axis=2):
    def delete_bond(a1, a2):
        bonds = set(a1.Bonds) & set(a2.Bonds)
        if bonds:
            print(bonds)
            for bond in bonds:
                atom3d.delete_bond(bond)

    atom3d.convert_to_in_cell()
    atom3d.clear_all_joints()
    cutoff = 10.0
    lo = atom3d.box_l[axis] + cutoff
    hi = atom3d.box_h[axis] - cutoff
    for atom_id in atom3d.Atoms.keys():
        atom = atom3d.Atoms[atom_id]
        if atom.coordinate[axis] > hi or atom.coordinate[axis] < lo:
            connects = atom3d.get_connected_atom_id(atom_id)
            for c_id in connects:
                catom = atom3d.Atoms[c_id]
                if (atom.coordinate[axis] > hi and catom.coordinate[axis] < lo) or (
                    catom.coordinate[axis] > hi and atom.coordinate[axis] < lo):
                    delete_bond(atom, catom)

    atom3d.sort_atom_bond_id()
    atom3d.cal_all_joints()

```
