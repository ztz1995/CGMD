```python
import numpy as np
import matplotlib.pyplot as plt
import pickle as pkl
import os, sys
sys.path.append("/home/centos/CGMD/")
# from pyMD import analysis as ans
config = {
    "mathtext.fontset":"stix",
    "font.family":'STIXGeneral',
    "font.serif":['SimSun'],
}
from matplotlib import rcParams
rcParams.update(config)

from matplotlib.font_manager import FontProperties
SimSun = FontProperties(fname='/home/centos/anaconda3/envs/md/lib/python3.7/site-packages/matplotlib/mpl-data/fonts/ttf/SimSun.ttf')
```

```python
inch_cm = 1
fig = plt.figure(dpi=400, figsize=(4 * inch_cm, 3 * inch_cm))

fontsize=12

labels = ["AA", "CGU", "CG"]
aver = [500, 500, 500]
msd = dict()

for i in range(3):
    with open("data/msd_cm_type_es_%s_%d.pkl" % (labels[i].lower(), aver[i]), "rb") as f:
        msd[labels[i]] = pkl.load(f)
_type = "hard"
_type2 = "soft"
markers = {"AA":"o", "CGU":"v", "CG":"^"}
colors = {"AA":"tab:red", "CGU":"tab:blue", "CG":"0.15"}
for i, label in enumerate(labels):
    plt.plot(msd["AA"]["time"], msd[label][_type2] / msd[label][_type], label=label, markerfacecolor='none',marker=markers[label], color=colors[label], zorder=3-i)
    m = (msd[label][_type2] / msd[label][_type]).max()
    m_id = (msd[label][_type2] / msd[label][_type]).argmax()
    print(label, msd["AA"]["time"][m_id], m)

plt.legend(fontsize=fontsize+2, frameon=False)
plt.xlabel(r'$t \ \mathbf{[} \mathrm{ns} \mathbf{]}$', fontdict={"fontproperties":SimSun, "fontsize":fontsize+4})
plt.ylabel(r'MSD 比率', fontdict={"fontproperties":SimSun, "fontsize":fontsize+4})
plt.xticks(fontsize=fontsize)
plt.yticks(fontsize=fontsize)

plt.xscale("log")
plt.yscale("log")
plt.ylim(bottom=0.5)
# plt.axis('equal')
plt.tight_layout()
plt.savefig("2 msd_4.jpg", dpi=400)
plt.show()
```
