```python
def customize_violin(parts, color=None, label=""):
    for key in parts:
        if key == "bodies":
            for i, pc in enumerate(parts[key]):
                if color is not None:
                    pc.set_facecolor(color)  # '#D43F3A'
                pc.set_alpha(0.7)
                pc.set_edgecolor('black')
                if i == 0:
                    pc.set_label(label)
                # pc.set_edgecolor('#D43F3A')
        else:
            parts[key].set_color('black')
```
